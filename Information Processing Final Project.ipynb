```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import csv
```

```python
original_data=pd.read_csv("train.csv")
```

```python
date_time=pd.DataFrame(original_data.Dates.str.split(' ',1).tolist(), columns = ['date','time'])
# here I am splitting the Dates column into the date and time
```

```python
date_time
# checking what the two columns look like
```

```python
rest_data=original_data.drop(columns='Dates')
#Taking the Dates column out to add date and time columns to the data.
```

```python
data=pd.concat([date_time,rest_data],axis=1)
#data has the processed columns and ready to be worked on
```

```python
data
```

```python
# getting the date column out
date_col=data[['date']]
date_col
```

```python
#takes the count of each unique value in the column and stores it in the list y
y=[]
for i in date_col.groupby('date', sort=False).size():
    y.append(i)
```

```python
#a view of the unique values of the date column
date_col.date.unique()
```

```python
#taking these unique values and saving them in list x
x=[]
for i in date_col.date.unique():
    x.append(i)
```

```python
#saving the x and y values in a new dataframe in their respective columns
date_table=pd.DataFrame({'Date':x,'Crimes':y}, columns=['Date','Crimes'])
date_table
```

```python
#creates a plot for the top 10 dates where the most crimes took place
stat_sorted = date_table.sort_values("Crimes", ascending=False)
ax = stat_sorted[:10].plot(x='Date', y="Crimes", kind='barh', figsize=(12,8), fontsize=15)
plt.title('Top 10 Dates of crime', fontsize=15)
plt.show()
```

```python
# getting the districts column out
district_col=data[['PdDistrict']]
district_col
```

```python
#a view of all the unique PD district values
district_col.PdDistrict.unique()
```

```python
#getting all those unique values of districts into a list called x
x=[]
for i in district_col.PdDistrict.unique():
    x.append(i)
```

```python
x
```

```python
#counting the occurences of the unique districts and saving the values in the list y
y=[]
for i in district_col.groupby('PdDistrict', sort=False).size():
    y.append(i)
```

```python
y
```

```python
#saving the x and y values in a new dataframe in their respective columns
district_table=pd.DataFrame({'PdDistrict':x,'Crimes':y}, columns=['PdDistrict','Crimes'])
district_table
```

```python
#creates a horizontal bar graph for crimes presented on district level
ax = district_table.plot(x='PdDistrict', y="Crimes", kind='barh', figsize=(12,5), fontsize=15)
plt.title('Crimes based on districts', fontsize=15)
plt.show()
```

```python
# Create a pie chart for the same data above
plt.pie(
    # using crimes in the data
    district_table['Crimes'],
    # with the labels being the police department districts
    labels=district_table['PdDistrict'],
    # with two slides exploded out, one for the maximum and one for the minimum
    explode=(0, 0, 0, 0, 0.15, 0, 0, 0, 0, 0.25),
    # with no shadows
    shadow=False,
    # with the start angle at 90%
    startangle=90,
    # with the percent listed as a fraction
    autopct='%1.1f%%',
    )

# View the plot drop above
plt.axis('equal')

# View the plot
plt.title('Crimes based on districts', fontsize=15)
plt.tight_layout()
plt.show()
```

```python
# getting the days of the week column out
day_col=data[['DayOfWeek']]
day_col
```

```python
# getting  view of the unique day values in the dataframe
day_col.DayOfWeek.unique()
```

```python
#saving the above unique values in a list called x 
x=[]
for i in day_col.DayOfWeek.unique():
    x.append(i)
```

```python
x
```

```python
#counting the the total of all these unique day values and saving the count in the list called y
y=[]
for i in day_col.groupby('DayOfWeek', sort=False).size():
    y.append(i)
```

```python
y
```

```python
#saving the x and y values in a new dataframe in their respective columns
day_table=pd.DataFrame({'DayOfWeek':x,'Crimes':y}, columns=['DayOfWeek','Crimes'])
day_table
```

```python
#creates a crime plot for the days of the week
ax = day_table.plot(x='DayOfWeek', y="Crimes", kind="line", linestyle=":", marker="D", figsize=(14,4), fontsize=15)
plt.xticks(day_table.index, day_table['DayOfWeek'])
plt.title('Top Days of crime', fontsize=15)
plt.show()
```

```python
# taking the different category of crimes out
category_col=data[['Category']]
category_col
```

```python
# a view of all the unique category values in the dataframe
category_col.Category.unique()
```

```python
#Taking the above unique values out in the list called x
x=[]
for i in category_col.Category.unique():
    x.append(i)
```

```python
#Counting the unique values in the category column and saving them in the list y
y=[]
for i in category_col.groupby('Category', sort=False).size():
    y.append(i)
```

```python
#saving the x and y values in a new dataframe in their respective columns
category_table=pd.DataFrame({'Category':x,'Crimes':y}, columns=['Category','Crimes'])
category_table
```

```python
#creates a crime plot for the various different categories of crime.
ax = category_table.plot(x='Category', y="Crimes", kind="line", linestyle="-", marker="o", figsize=(14,4), fontsize=10)
plt.xticks(category_table.index, category_table['Category'], rotation=90)
plt.title('Categories of crimes and their occurances', fontsize=15)
plt.show()
```
